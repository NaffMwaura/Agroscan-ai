```python
# # # # # import pandas as pd
# # # # # import matplotlib.pyplot as plt
# # # # # import seaborn as sns



# # # # # dt1 = pd.read_csv("/content/student-por.csv")
# # # # # dt2 = pd.read_csv("/content/student-mat.csv")

# # # # # dt1 = pd.read_csv("/content/student-por.csv", sep=';')
# # # # # dt2 = pd.read_csv("/content/student-mat.csv", sep=';')

# # # # # print(dt1.head())
# # # # # print(dt2.head())
# # # # # print(dt1.describe())

# # # # # print(dt1.info())
# # # # # print(dt2.info())
# # # # # print(dt1.isnull().sum())
# # # # # print(dt2.isnull().sum())
# # # # # mean_g3_por = dt1["G3"].mean()
# # # # # mean_g3_mat = dt2["G3"].mean()

# # # # # print("Mean Final Grade in Portuguese Class:", mean_g3_por)
# # # # # print("Mean Final Grade in Math Class:", mean_g3_mat)

# # # # # plt.hist(dt1["G3"], bins=10, color='skyblue', edgecolor='black')
# # # # # plt.title("Distribution of Final Grade (G3) - Portuguese Class")
# # # # # plt.xlabel("Grade")
# # # # # plt.ylabel("Number of Students")
# # # # # plt.grid(True)
# # # # # plt.show()

# # # # # sns.histplot(data=dt1, x="G3", kde=True, color="green", bins=10)
# # # # # plt.title("Final Grade Distribution (G3) - Portuguese Class")
# # # # # plt.xlabel("Final Grade")
# # # # # plt.ylabel("Student Count")
# # # # # plt.show()
# # # # # dt_all = pd.concat([dt1, dt2], ignore_index=True)

# # # # # sns.boxplot(x="studytime", y="G3", data=dt_all)

# # # # # plt.title("Distribution of Final Grade by Study Time")
# # # # # plt.xlabel("My study Time")
# # # # # plt.ylabel("Final Grade (G3)")
# # # # # plt.show()

# # # # # sns.countplot(x='sex', data=dt_all)

# # # # # plt.title("Number of Students by Gender")
# # # # # plt.xlabel("Gender")
# # # # # plt.ylabel("Count")
# # # # # plt.show()

# # # # # # num_cols = ['age', 'absences', 'G1', 'G2', 'G3']

# # # # # # sns.pairplot(dt_all[num_cols])

# # # # # # plt.suptitle("Pairwise Relationships Between Selected Numeric Variables", y=1.02)
# # # # # # plt.show()

# # # # # corr = dt_all.corr(numeric_only=True)

# # # # # plt.figure(figsize=(10, 8))
# # # # # sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")

# # # # # plt.title("Correlation Heatmap")
# # # # # plt.show()

# # # # import pandas as pd
# # # # df1 = pd.read_csv("/winequality-red.csv",sep=';')
# # # # df2 = pd.read_csv("/winequality-white.csv",sep=';')
# # # # df1.head()
# # # # df2.head()
# # # # df_all = pd.merge(df1,df2)
# # # # df_all.head()
# # # # df_all = pd.concat([df1,df2],axis=0)
# # # # df_all.head()
# # # # df_all.isnull().sum()
# # # # df_all.drop_duplicates(inplace=True)
# # # # df_all.info()
# # # # df_all.describe()
# # # # import seaborn as sns
# # # # import matplotlib.pyplot as plt
# # # # sns.histplot(df_all['quality'], kde=True)
# # # # plt.title("Distribution of Wine Quality")
# # # # plt.show()
# # # # sns.boxplot(x='quality', y='alcohol', data=df_all)
# # # # plt.title("Boxplot: Alcohol Content vs Wine Quality")
# # # # plt.show()
# # # # sns.violinplot(x='quality', y='alcohol', data=df_all)
# # # # plt.title("Violinplot: Alcohol Content vs Wine Quality")
# # # # plt.show()
# # # # sns.violinplot(x="quality", y="alcohol", data=df_all)
# # # # plt.title("Violinplot: Alcohol Content vs Wine Quality")
# # # # plt.show()
# # # # sns.histplot(x='quality', data=df_all)
# # # # plt.title("Distribution of Wine Types (Red vs White)")
# # # # plt.show()


# # # from sklearn.metrics import roc_curve, auc, roc_auc_score
# # # from sklearn.preprocessing import label_binarize
# # # import matplotlib.pyplot as plt
# # # import numpy as np

# # # # Binarize the output for multi-class
# # # y_test_bin = label_binarize(y_test, classes=[0, 1, 2])

# # # # Number of classes
# # # n_classes = len(class_names)

# # # # Plotting
# # # plt.figure()
# # # colors = ['darkorange', 'darkgreen', 'darkblue']

# # # for i in range(n_classes):
# # #     # compute ROC curve
# # #     fpr, tpr, _ = roc_curve(y_test_bin[:, i],
# # #                             models['SVM'].decision_function(X_test_scaled)[:, i]
# # #                             if hasattr(models['SVM'], "decision_function")
# # #                             else models['SVM'].predict_proba(X_test_scaled)[:, i],
# # #                             pos_label=1)
# # #     roc_auc = auc(fpr, tpr)
# # #     plt.plot(fpr, tpr, color=colors[i],
# # #               label='Class {0} (AUC = {1:0.2f})'.format(class_names[i], roc_auc))


# # # plt.plot([0, 1], [0, 1], 'k--', label='Random Guess')
# # # plt.xlabel('False Positive Rate')
# # # plt.ylabel('True Positive Rate')
# # # plt.title('ROC Curve for Iris SVM')
# # # plt.legend(loc='lower right')
# # # plt.grid()
# # # plt.show()

# # from sklearn.metrics import roc_curve, auc, roc_auc_score
# # from sklearn.preprocessing import label_binarize
# # import matplotlib.pyplot as plt
# # import numpy as np

# # # Binarize the output
# # y_test_bin = label_binarize(y_test, classes=[0, 1, 2])

# # # Number of classes
# # n_classes = len(class_names)

# # # Store FPR, TPR, and AUC for each class
# # fpr = {}
# # tpr = {}
# # roc_auc = {}

# # # Calculate for each class
# # for i in range(n_classes):
# #     fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i],
# #                                    models['SVM'].decision_function(X_test_scaled)[:, i],
# #                                    pos_label=1)
# #     roc_auc[i] = auc(fpr[i], tpr[i])

# # # Micro-average
# # fpr["micro"], tpr["micro"], _ = roc_curve(y_test_bin.ravel(),
# #                                          models['SVM'].decision_function(X_test_scaled).ravel())
# # roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

# # # Macro-average
# # all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))
# # mean_tpr = np.zeros_like(all_fpr)
# # for i in range(n_classes):
# #     mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])

# # mean_tpr /= n_classes
# # fpr["macro"] = all_fpr
# # tpr["macro"] = mean_tpr
# # roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])

# # # Plotting
# # plt.figure()
# # colors = ['darkorange', 'darkgreen', 'darkblue']

# # for i in range(n_classes):
# #     plt.plot(fpr[i], tpr[i],
# #               color=colors[i],
# #               label='Class {0} (AUC = {1:0.2f})'.format(class_names[i], roc_auc[i]))

# # plt.plot(fpr["micro"], tpr["micro"], label='Micro-average (AUC = {0:0.2f})'.format(roc_auc["micro"]), color='deeppink', linestyle='--')
# # plt.plot(fpr["macro"], tpr["macro"], label='Macro-average (AUC = {0:0.2f})'.format(roc_auc["macro"]), color='navy', linestyle='--')

# # plt.plot([0, 1], [0, 1], 'k--', label='Random Guess')

# # plt.xlabel('False Positive Rate')
# # plt.ylabel('True Positive Rate')
# # plt.title('Multi-class ROC with Micro and Macro Average')
# # plt.legend(loc='lower right')
# # plt.grid()
# # plt.show()

# # ML Libraries
# from sklearn.model_selection import train_test_split
# from sklearn.preprocessing import StandardScaler, label_binarize
# from sklearn.tree import DecisionTreeClassifier
# from sklearn.ensemble import RandomForestClassifier
# from sklearn.svm import SVC
# from sklearn.metrics import accuracy_score, classification_report, roc_curve, auc, confusion_matrix

# # Data Manipulation Libraries
# import pandas as pd
# import matplotlib.pyplot as plt
# import seaborn as sns
# import numpy as np

# # File upload for Google Colab
# from google.colab import files

# # ------------------------------------------------------------------------------
# # 1️⃣ Upload CSV File
# # ------------------------------------------------------------------------------
# print("Please upload your CSV file")
# uploaded = files.upload()
# filename = list(uploaded.keys())[0]

# # ------------------------------------------------------------------------------
# # 2️⃣ Loading the CSV into a DataFrame
# # ------------------------------------------------------------------------------
# df = pd.read_csv(filename)

# # ------------------------------------------------------------------------------
# # 3️⃣ Split into Features (X) and Target (y)
# # ------------------------------------------------------------------------------
# # Change 'your_label_column_name' to your own label column
# label = 'your_label_column_name'

# X = df.drop([label], axis=1)
# y = df[label]

# # ------------------------------------------------------------------------------
# # 4️⃣ Split into Train and Test sets
# # ------------------------------------------------------------------------------
# X_train, X_test, y_train, y_test = train_test_split(
#     X, y, test_size=0.3, random_state=42
# )

# # ------------------------------------------------------------------------------
# # 5️⃣ Standardize (for SVM)
# # ------------------------------------------------------------------------------
# scaler = StandardScaler()
# X_train_scaled = scaler.fit_transform(X_train)
# X_test_scaled = scaler.transform(X_test)

# # ------------------------------------------------------------------------------
# # 6️⃣ Define models to train
# # ------------------------------------------------------------------------------
# models = {
#     "Decision Tree": DecisionTreeClassifier(random_state=42),
#     "Random Forest": RandomForestClassifier(random_state=42),
#     "SVM": SVC(random_state=42, probability=True)
# }

# # ------------------------------------------------------------------------------
# # 7️⃣ Train and evaluate models
# # ------------------------------------------------------------------------------
# for name, model in models.items():
#     print(f"Training: {name}")

#     if name == "SVM":
#         model.fit(X_train_scaled, y_train)
#         y_pred = model.predict(X_test_scaled)
#         y_pred_proba = model.predict_proba(X_test_scaled)
#     else:
#         model.fit(X_train, y_train)
#         y_pred = model.predict(X_test)
#         y_pred_proba = model.predict_proba(X_test)

#     accuracy = accuracy_score(y_test, y_pred)
#     print(f"Accuracy for {name}: {accuracy * 100:.2f}%")
#     print("Classification Report:")
#     print(classification_report(y_test, y_pred))

#     print("-"*50)

# # ------------------------------------------------------------------------------
# # 8️⃣ Multi-class ROC Curve (SVM)
# # ------------------------------------------------------------------------------
# # Perform label binarization first
# classes = np.unique(y)
# y_test_bin = label_binarize(y_test, classes=classes)
# n_classes = len(classes)

# # Calculate ROC for each class
# fpr = {}
# tpr = {}
# roc_auc = {}

# model = models['SVM']

# # Use decision scores or predict_proba
# scores = model.decision_function(X_test_scaled) if hasattr(model, "decision_function") else model.predict_proba(X_test_scaled)

# for i in range(n_classes):
#     fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], scores[:, i], pos_label=1)
#     roc_auc[i] = auc(fpr[i], tpr[i])

# # Micro-average
# fpr["micro"], tpr["micro"], _ = roc_curve(y_test_bin.ravel(), scores.ravel())
# roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

# # Macro-average
# all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))
# mean_tpr = np.zeros_like(all_fpr)
# for i in range(n_classes):
#     mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])

# mean_tpr /= n_classes

# fpr["macro"] = all_fpr
# tpr["macro"] = mean_tpr
# roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])

# # Plotting
# plt.figure()
# colors = ['darkorange', 'darkgreen', 'darkblue']

# for i in range(n_classes):
#     plt.plot(fpr[i], tpr[i],
#               color=colors[i],
#               label='Class {0} (AUC = {1:0.2f})'.format(classes[i], roc_auc[i]))

# plt.plot(fpr["micro"], tpr["micro"], label='Micro-average (AUC = {0:0.2f})'.format(roc_auc["micro"]), color='deeppink', linestyle='--')
# plt.plot(fpr["macro"], tpr["macro"], label='Macro-average (AUC = {0:0.2f})'.format(roc_auc["macro"]), color='navy', linestyle='--')

# plt.plot([0, 1], [0, 1], 'k--', label='Random Guess')

# plt.xlabel('False Positive Rate')
# plt.ylabel('True Positive Rate')
# plt.title('Multi-class ROC with Micro and Macro Average')
# plt.legend(loc='lower right')
# plt.grid()
# plt.show()
# ML Libraries
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, roc_curve, auc, confusion_matrix

# Data Manipulation Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# File upload for Google Colab
from google.colab import files

# ------------------------------------------------------------------------------
# 1️⃣ Upload CSV files
# ------------------------------------------------------------------------------
print("Please upload your training CSV first")
uploaded = files.upload()
train_file = list(uploaded.keys())[0]

print("Please upload your testing CSV now")
uploaded = files.upload()
test_file = list(uploaded.keys())[0]

# ------------------------------------------------------------------------------
# 2️⃣ Loading the CSVs into DataFrames
# ------------------------------------------------------------------------------
train_df = pd.read_csv(train_file)
test_df = pd.read_csv(test_file)

# ------------------------------------------------------------------------------
# 3️⃣ Split into Features (X) and Target (y)
# ------------------------------------------------------------------------------
# Change 'your_label_column_name' to your label column
label = 'your_label_column_name'

X_train = train_df.drop([label], axis=1)
y_train = train_df[label]

X_test = test_df.drop([label], axis=1)
y_test = test_df[label]

# ------------------------------------------------------------------------------
# 4️⃣ Standardize (for SVM)
# ------------------------------------------------------------------------------
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ------------------------------------------------------------------------------
# 5️⃣ Define models to train
# ------------------------------------------------------------------------------
models = {
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "SVM": SVC(random_state=42, probability=True)
}

# ------------------------------------------------------------------------------
# 6️⃣ Train and evaluate models
# ------------------------------------------------------------------------------
for name, model in models.items():
    print(f"Training: {name}")

    if name == "SVM":
        model.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_test_scaled)
        y_pred_proba = model.predict_proba(X_test_scaled)
    else:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        y_pred_proba = model.predict_proba(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy for {name}: {accuracy * 100:.2f}%")
    print("Classification Report:")
    print(classification_report(y_test, y_pred))

    print("-"*50)

# ------------------------------------------------------------------------------
# 7️⃣ Multi-class ROC Curve (SVM)
# ------------------------------------------------------------------------------
# Perform label binarization first
classes = np.unique(y_test)
y_test_bin = label_binarize(y_test, classes=classes)
n_classes = len(classes)

model = models['SVM']

# Use decision scores or predict_proba
scores = model.decision_function(X_test_scaled) if hasattr(model, "decision_function") else model.predict_proba(X_test_scaled)

fpr = {}
tpr = {}
roc_auc = {}

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], scores[:, i], pos_label=1)
    roc_auc[i] = auc(fpr[i], tpr[i])

# Micro-average
fpr["micro"], tpr["micro"], _ = roc_curve(y_test_bin.ravel(), scores.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

# Macro-average
all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))
mean_tpr = np.zeros_like(all_fpr)
for i in range(n_classes):
    mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])

mean_tpr /= n_classes

fpr["macro"] = all_fpr
tpr["macro"] = mean_tpr
roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])

# Plotting
plt.figure()
colors = ['darkorange', 'darkgreen', 'darkblue']

for i in range(n_classes):
    plt.plot(fpr[i], tpr[i],
              color=colors[i],
              label='Class {0} (AUC = {1:0.2f})'.format(classes[i], roc_auc[i]))

plt.plot(fpr["micro"], tpr["micro"], label='Micro-average (AUC = {0:0.2f})'.format(roc_auc["micro"]), color='deeppink', linestyle='--')
plt.plot(fpr["macro"], tpr["macro"], label='Macro-average (AUC = {0:0.2f})'.format(roc_auc["macro"]), color='navy', linestyle='--')

plt.plot([0, 1], [0, 1], 'k--', label='Random Guess')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multi-class ROC with Micro and Macro Average')
plt.legend(loc='lower right')
plt.grid()
plt.show()

```
